```python
#Add Pandas as Dependency
import pandas as pd
import os
import csv
```

```python
#Files to Load
schools_to_load = "Resources/schools_complete.csv"
students_to_load = "Resources/students_complete.csv"
```

```python
#Read the files
school_data_df = pd.read_csv(schools_to_load)
school_data_df
```

```python
#Read the Student Data
students_data_df = pd.read_csv(students_to_load)
students_data_df
```

```python
students_data_df.count()
```

```python
students_data_df.notnull().sum()

```

```python
#Determine Student Data Types
students_data_df.dtypes
```

```python
#Create prefix and suffix list
prefixes_suffixes = ("Dr. ", "Mr. ", "Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " PhD", " DVM")

#Iterate through pre/suf list
for word in prefixes_suffixes:
    students_data_df["student_name"] = students_data_df["student_name"].str.replace(word,"")

students_data_df
```

```python
#Merge two data frames
school_data_complete_df = pd.merge(students_data_df, school_data_df, on=["school_name", "school_name"])
school_data_complete_df
```

```python
#Number of Students
student_count_a = school_data_complete_df["Student ID"].count()
student_count_a
```

```python
school_data_complete_df["Student ID"].count()
```

```python
student_count_a.dtype
```

```python
# Calculate the total number of schools
school_count_2 = len(school_data_complete_df["school_name"].unique())
school_count_2
```

```python
#Total budget
school_budget_2 = sum(school_data_df["budget"])
school_budget_2
```

```python
#Student Math Averages
math_average = school_data_complete_df["math_score"].mean()
math_average
```

```python
#Student Reading Averages
reading_average = school_data_complete_df["reading_score"].mean()
reading_average
```

```python
#Set Passing Scores
passing_math_a = school_data_complete_df["math_score"]>=70
passing_reading_a = school_data_complete_df["reading_score"]>=70
```

```python
passing_math_a
```

```python
passing_reading_a
```

```python
passing_math = school_data_complete_df[school_data_complete_df["math_score"]>=70]
passing_math.head()
```

```python
passing_reading = school_data_complete_df[school_data_complete_df["reading_score"]>=70]
passing_reading.head()
```

```python
#Number of Students who passed Math & Reading
passing_math_count = passing_math["Student ID"].count()
passing_math_count
```

```python

passing_reading_count = passing_reading["Student ID"].count()
passing_reading_count
```

```python
#Percent that Passed Math
passing_math_percentage = passing_math_count / (student_count_a) *100
print(passing_math_percentage)


#Percent that Passed Reading
passing_reading_percentage = passing_reading_count / (student_count_a) *100
print(passing_reading_percentage)
```

```python
# Calculate the students who passed both math and reading.
passing_math_reading = school_data_complete_df[(school_data_complete_df["math_score"] >= 70) & (school_data_complete_df["reading_score"] >= 70)]

passing_math_reading.head()
```

```python
# Calculate the number of students who passed both math and reading.
overall_passing_math_reading_count = passing_math_reading["Student ID"].count()
overall_passing_math_reading_count
```

```python
## Calculate the overall passing percentage.
overall_passing_percentage = overall_passing_math_reading_count / student_count_a * 100
overall_passing_percentage
```

```python
#Add new columns for district Summary
district_summary_df = pd.DataFrame(
          [{"Total Schools": school_count_2,
          "Total Students": student_count_a,
          "Total Budget": school_budget_2,
          "Average Math Score": math_average,
          "Average Reading Score": reading_average,
          "% Passing Math": passing_math_percentage,
         "% Passing Reading": passing_reading_percentage,
        "% Overall Passing": overall_passing_percentage}])
district_summary_df
```

```python
district_summary_df.dtypes
```

```python
# Format the Columns
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.1f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.1f}".format)
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:.0f}".format)
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:.0f}".format)
district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].map("{:.0f}".format)                                                                                      

district_summary_df
```
